```python
#necessary import modules
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from data import *
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#structure of artificial neural network 
retinal_position = keras.Input(shape=(81,), name='retinal position')
eye_position = keras.Input(shape=(81,), name='eye position')
combined = layers.concatenate([retinal_position, eye_position])
dense = layers.Dense(30, activation='sigmoid', name='hidden')(combined)
output = layers.Dense(81, activation='sigmoid', name='output')(dense)
model = keras.Model(inputs=[retinal_position, eye_position], outputs=[output])
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
#visual representation of summary 
model.summary()
```

```python
#directed graph of structure of network 
keras.utils.plot_model(model, show_shapes=True, show_layer_names=True, to_file='model.png')
Image('model.png')
```

```python
#training of artificial neural network over 500 epochs in batches of 100
history = model.fit({'retinal position': retinal_data_training, 'eye position': eye_data_training}, 
                    {'output': targets_training}, 
                    epochs=500, 
                    batch_size=100, 
                    validation_data=({'retinal position': retinal_data_test, 'eye position': eye_data_test}, 
                                     {'output': targets_test})
                   )
```

```python
#graph of training and valdiation loss
history_dict = history.history
loss_values = history_dict['loss']
epochs = range(1, len(loss_values) + 1)
val_loss_values = history_dict['val_loss']
plt.plot(epochs, loss_values, 'ro', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#graph of training and validation accuracy 
acc = history_dict['acc']
val_acc = history_dict['val_acc']
plt.plot(epochs, acc, 'ro', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#confirmation that network is able to accurately predict test data
results = model.evaluate({'retinal position': retinal_data_test, 'eye position': eye_data_test}, {'output': targets_test})
```

```python
#predicted output of a single datum from test data 
prediction = model.predict({'retinal position': retinal_data_test, 'eye position': eye_data_test})
print(prediction[1])
```
